```python
import numpy as np
import matplotlib.pyplot as plt

def uniform_distribution(a, b, n):
    return np.random.uniform(a, b, n)

# Take input from the user
a = float(input("Enter the lower bound (a): "))
b = float(input("Enter the upper bound (b): "))
n = int(input("Enter the number of samples (n): ")) 


# Generate the random samples
samples = uniform_distribution(a, b, n)

# Print the samples
print("Samples:")
print(samples)

# Calculate the mean and variance
mean = np.mean(samples)
variance = np.var(samples)

print(f"\nMean: {mean:.2f}")
print(f"Variance: {variance:.2f}")

# Plot the histogram
plt.figure(figsize=(5, 3))
plt.hist(samples, bins=20, density=True)
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.title(f"Uniform Distribution (a={a}, b={b})")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

# Parameters for the uniform distribution
a = float(input("Enter the lower bound (a): "))
b = float(input("Enter the upper bound (b): "))

# Generate data points
x_values = np.linspace(a - 1, b + 1, 1000)  # Extend range slightly for better plot visualization

# Calculate the PDF values for each x
pdf_values = uniform.pdf(x_values, loc=a, scale=b-a)

# Print the probability table for selected points
print("x\tP(X = x)")
for x in np.linspace(a, b, 10):
    print(f"{x:.2f}\t{uniform.pdf(x, loc=a, scale=b-a):.4f}")

# Plot the uniform distribution
plt.figure(figsize=(8, 5))
plt.plot(x_values, pdf_values, label=f'Uniform Distribution\n$a={a}$, $b={b}$', color='skyblue')
plt.fill_between(x_values, pdf_values, alpha=0.2, color='skyblue')
plt.xlabel('x')
plt.ylabel('Probability Density Function (f(x))')
plt.title('Uniform Probability Distribution')
plt.legend()
plt.grid(True)
plt.show()
```
