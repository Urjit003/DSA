```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import hypergeom

# Parameters for Hypergeometric distribution
N = int(input("Enter the population size (N): "))
K = int(input("Enter the number of successes in the population (K): "))
n = int(input("Enter the number of draws (n): "))
max_k = min(K, n)  # Maximum number of successes to consider

# Calculate PMF values for each k from 0 to max_k
k_values = np.arange(0, max_k + 1)
hypergeom_pmf = hypergeom.pmf(k_values, N, K, n)

# Print table of k vs P(X = k)
print("k\tP(X = k)")
for k in range(max_k + 1):
    print(f"{k}\t{hypergeom_pmf[k]:.4f}")

# Plot the Hypergeometric distribution
plt.figure(figsize=(8, 5))
plt.bar(k_values, hypergeom_pmf, color='skyblue')
plt.xlabel("Number of successes (k)")
plt.ylabel("Probability mass function (P(X = k))")
plt.title(f"Hypergeometric Distribution (N={N}, K={K}, n={n})")
plt.grid(True)
plt.show()
```
