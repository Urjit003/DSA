```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom
```

```python
n = int(input("Enter the number of trials (n): "))
p = float(input("Enter the probability of success (p): "))
```

```python
x_values = np.arange(0, n + 1)
f_x = binom.pmf(x_values, n, p)

# Print table of x vs P(X = x)
print("X\tP(X)")
for i in range(n + 1):
    print(f"{x_values[i]}\t{f_x[i]:.4f}")
```

```python
plt.figure(figsize=(8, 5))
plt.bar(x_values, f_x, color='skyblue')
plt.xlabel("Number of successes (x)")
plt.ylabel("Probability mass function (f(x))")
plt.title(f"Binomial Distribution (n={n}, p={p})")
plt.xticks(x_values)
plt.grid(True)
plt.show()
```
