```python
import numpy as np
import matplotlib.pyplot as plt
import math

# Hardcoded parameters for normal distribution
mu = 0      # Mean
sigma = 1   # Standard deviation
n = 1000    # Number of data points

# Generate data points
x = np.linspace(mu - 4*sigma, mu + 4*sigma, n)

# Calculate the PDF of the normal distribution
def normal_pdf(x, mu, sigma):
    return (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma) ** 2)

# Calculate the standard deviation (hardcoded formula)
def calculate_standard_deviation(data, mean):
    variance = sum((x - mean) ** 2 for x in data) / len(data)
    return math.sqrt(variance)

# Calculate PDF values
pdf_values = normal_pdf(x, mu, sigma)

# Print the standard deviation
calculated_std_dev = calculate_standard_deviation(x, mu)
print(f"Calculated Standard Deviation: {calculated_std_dev:.4f}")

# Plot the normal distribution
plt.figure(figsize=(8, 5))
plt.plot(x, pdf_values, label=f'Normal Distribution\n$\mu={mu}$, $\sigma={sigma}$')
plt.xlabel('x')
plt.ylabel('Probability Density Function (f(x))')
plt.title('Normal Probability Distribution')
plt.legend()
plt.grid(True)
plt.show()
```
