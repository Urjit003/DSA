```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parameters for Poisson distribution
lambda_ = float(input("Enter the average rate of occurrence (Î»): "))
max_k = 20  # Maximum number of events to consider

# Calculate PMF values for each k from 0 to max_k
k_values = np.arange(0, max_k + 1)
poisson_pmf = poisson.pmf(k_values, lambda_)

# Print table of k vs P(X = k)
print("k\tP(X = k)")
for k in range(max_k + 1):
    print(f"{k}\t{poisson_pmf[k]:.4f}")

# Plot the Poisson distribution
plt.figure(figsize=(8, 5))
plt.bar(k_values, poisson_pmf, color='skyblue')
plt.xlabel("Number of events (k)")
plt.ylabel("Probability mass function (P(X = k))")
plt.title(f"Poisson Distribution ($\lambda={lambda_})")
plt.grid(True)
plt.show()
```
